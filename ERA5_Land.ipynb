Iterative importing of 4 months per site for variable investigation in September to December 2014

Takes roughly 5 hours

```python
final_results = []

start = time.time()
for site in sites:
    s_start = time.time()
    for month in months:
        m_start = time.time()
        file_path = f'{site}_{month}.grib'
        
        grbs = pygrib.open(file_path)
        era5_list = []
        for var in selected_vars:
            msgs = grbs.select(name=var)
            for msg in msgs:
                data = msg.values.flatten()  # Flatten the 2D array to 1D
                lats, lons = msg.latlons()
                date = msg.analDate
                formatted_date = datetime.strftime(date, '%Y-%m-%d') 
                level = msg.level
                era5_list.append({
                    'variable': var,
                    'data': data.mean(),  # Calculate mean of the flattened array
                    'latitude': lats.flatten(),  # Flatten the latitudes
                    'longitude': lons.flatten(),  # Flatten the longitudes
                    'date': formatted_date,
                    'level': level
                })

        era5_df = pd.DataFrame(era5_list)

        era5 = era5_df.pivot_table(
            index='date', 
            columns='variable', 
            values='data', 
            aggfunc='mean'
        )

        era5 = era5.reset_index().rename_axis(None, axis=1)

        era5 = era5.rename(columns=rename_dict)

        for col, func in conversion_dict.items():
            if col in era5.columns:
                era5[col] = era5[col].apply(func)
        
        era5['Longitude'] = centres[site][0]
        era5['Latitude'] = centres[site][1]

        final_results.append(era5)
        m_end = time.time()
        m_time = m_end - m_start
        #print(f'Completed: {month}, {site}. Time taken: {round(m_time,2)}')
        
    s_end = time.time()
    s_time = s_end - s_start
    print(f'Time taken for {site}: {round(s_time,2)} seconds')

# Concatenate all the results
final_df = pd.concat(final_results, ignore_index=True)
end = time.time()
total_time = end - start
print(f'Overall time taken: {round(total_time,2)} seconds')
```

Iterative importing takes 12 hours

```python
valid_results = []
failed_imports = [] 

start = time.time()
for site in sites:
    s_start = time.time()
    for year in years:
        y_start = time.time()
        for month in months:
            m_start = time.time()
            file_path = f'{site}_{month}_{year}.grib'  # File path
            
            try:
                grbs = pygrib.open(file_path)
                valid_list = []
                lats, lons = None, None
                
                for var in selected_vars:
                    msgs = grbs.select(name=var)
                    for msg in msgs:
                        if lats is None or lons is None:
                            lats, lons = msg.latlons()  # Extract lat and lon
                        data = msg.values.flatten()  # Flatten the 2D array to 1D
                        date = msg.analDate
                        formatted_date = datetime.strftime(date, '%Y-%m-%d')
                        level = msg.level
                        valid_list.append({
                            'variable': var,
                            'data': data.mean(),  
                            'latitude': lats.flatten(), 
                            'longitude': lons.flatten(),  
                            'date': formatted_date,
                            'level': level
                        })

                valid_df = pd.DataFrame(valid_list)

                valid = valid_df.pivot_table(
                    index='date',
                    columns='variable',
                    values='data',
                    aggfunc='mean'
                )

                valid = valid.reset_index().rename_axis(None, axis=1)

                valid = valid.rename(columns=rename_dict)

                for col, func in conversion_dict.items():
                    if col in valid.columns:
                        valid[col] = valid[col].apply(func)

                # Assign latitude and longitude from site_info
                valid['Longitude'] = site_info[site]['Longitude']
                valid['Latitude'] = site_info[site]['Latitude']

                valid_results.append(valid)
            except Exception as e:
                print(f'Failed to process {file_path}: {str(e)}')
                failed_imports.append(file_path) # Error in importing capture

            m_end = time.time()
            m_time = m_end - m_start
            # print(f'Completed: {month}, {site}. Time taken: {round(m_time, 2)} seconds')

        y_end = time.time()
        y_time = round(y_end - y_start, 2)
        print(f'Time taken for {site, year}: {round(y_time, 2)} seconds')

    s_end = time.time()
    s_time = s_end - s_start
    print(f'Time taken for {site}: {round(s_time, 2)} seconds')

if valid_results:
    final_valid = pd.concat(valid_results, ignore_index=True)

end = time.time()
total_time = end - start
print(f'Overall time taken: {round(total_time, 2)} seconds')
```
